```python
import gzip
import pickle
import numpy as np

def load_data():
    f = gzip.open('mnist.pkl.gz','rb')
    (r,v,s) = pickle.load(f,encoding = 'iso-8859-1')
    (training_data,validation_data,test_data) = (r,v,s)
    f.close()
    
    return (training_data,validation_data,test_data)

mnist = load_data()
```
